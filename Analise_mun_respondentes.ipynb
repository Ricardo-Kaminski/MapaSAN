```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import openpyxl
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
import plotly
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
pd.set_option("display.precision", 2)
```

```python
file = "Municipal\Respostas\Formulario_2.xlsx"
df = pd.read_excel(file, header = 0)
df
```

```python
df.dtypes
```

```python
df = df.drop(labels=['versao_salva'], axis=1)
df = df.drop(labels=['data_hora_extracao'], axis=1)
```

```python
df
```

```python
table = pd.pivot_table(df, values = 'nome_municipio', columns=['uf'], aggfunc=len)
table_mun_uf = pd.DataFrame(table.T)
table_mun_uf.reset_index(inplace=True)
table_mun_uf = table_mun_uf.rename(columns = {'index':'new column name'})
table_mun_uf.columns = ['uf', 'mun_resp']
table_mun_uf
```

```python
total_mun_resp = table_mun_uf.mun_resp.sum()
```

```python
table_mun_uf['percentual'] = (table_mun_uf['mun_resp']/total_mun_resp)*100
table_mun_uf
```

```python
table_mun_uf = table_mun_uf.sort_values(by='percentual', ascending=False).reset_index(drop=True)
```

```python
table_mun_uf
```

```python
file = './estimativa_dou_2021.xls'
df_mun = pd.read_excel(file, header=0)
df_mun
```

```python
table = pd.pivot_table(df_mun, values = 'NOME DO MUNICÍPIO', columns=['UF'], aggfunc=len)
table_mun = pd.DataFrame(table.T)
table_mun.reset_index(inplace=True)
# table_mun = table_mun_uf.rename(columns = {'index':'new column name'})
table_mun.columns = ['uf', 'mun_mun']
table_mun
```

```python
mun = table_mun_uf.merge(table_mun, right_on= 'uf', left_on = 'uf')
mun.columns = ['uf', 'mun_resp', 'perc_resp', 'tot_mun']
mun
```

```python
total_mun_uf = mun['tot_mun']
mun['perc_tot'] = (table_mun_uf['mun_resp']/total_mun_uf)*100
mun
```

```python
mun = mun.sort_values(by='perc_tot', ascending=False).reset_index(drop=True)
mun
```

```python
# Plot de gráfico de barras horizontais municípios respondentes por UF
fig = px.bar(mun, 
             y ='mun_resp', 
             x = 'uf',
             # orientation='h', 
             title= 'Quantidade de municípios respondentes por UF',
             width=1200, 
             height=600,
             text_auto=True
             )
fig.update_layout(xaxis = {'categoryorder':'total descending'}) 
fig.show()
```
