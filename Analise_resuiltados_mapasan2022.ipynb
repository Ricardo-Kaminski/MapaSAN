```python
file = "estimativa_dou_2021.xls"
mun = pd.read_excel(file, sheet_name='Municípios')
mun
```

```python
mun_uf= mun['UF'].value_counts()
mun_uf = pd.DataFrame(mun_uf)
mun_uf.columns = ['tot_mun']
mun_uf
```

```python
mun_resp = df_uf.merge(mun_uf, left_index=True, right_index=True)
mun_resp.columns = ['mun_resp', 'perc_resp', 'mun_tot']
mun_resp['perc_tot'] = (mun_resp['mun_resp']/mun_resp[ 'mun_tot'])*100
mun_resp
```
